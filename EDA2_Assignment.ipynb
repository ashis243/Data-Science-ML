```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Standard Scaling
scaler_standard = StandardScaler()
df_standard_scaled = df.copy()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df_standard_scaled[numerical_cols] = scaler_standard.fit_transform(df[numerical_cols])
```

```python
# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[numerical_cols] = scaler_minmax.fit_transform(df[numerical_cols])
```

```python
print("Standard Scaled Data Sample:\n", df_standard_scaled.head())
print("Min-Max Scaled Data Sample:\n", df_minmax_scaled.head())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
# Encode each categorical column
for column in ['workclass','education','marital_status','occupation','relationship','race','native_country','income']:
    if column in df.columns:
        df[column] = label_encoder.fit_transform(df[column].astype(str))
```

```python
df
```

```python
df.head()
```

```python

# One-Hot Encoding for categorical variables with < 5 categories
df_encoded = pd.get_dummies(df, columns=['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country'], drop_first=True)
```

```python
df_encoded['income'] = label_encoder.fit_transform(df_encoded['income'])
```

```python
#Feature Engineering
# Example 1: Age Group
df_encoded['age_group'] = pd.cut(df_encoded['age'], bins=[0, 30, 50, 70, 100], labels=['0-30', '30-50', '50-70', '70+'])

# Example 2: Total Capital Gain/Loss
df_encoded['total_capital'] = df_encoded['capital_gain'] - df_encoded['capital_loss']
```

```python
df_encoded[['age_group', 'total_capital']]
```

```python
# Apply Transformation
import numpy as np
```

```python
# Example: Log transformation for capital_gain (if skewed)
df_encoded['log_capital_gain'] = np.log1p(df_encoded['capital_gain'])
```

```python
#Feature Selection
from sklearn.ensemble import IsolationForest
```

```python
# Initialize Isolation Forest
iso_forest = IsolationForest(contamination=0.01)  # Assume 1% outliers
df_encoded['outlier'] = iso_forest.fit_predict(df_encoded[numerical_cols])
```

```python
# Remove outliers
df_no_outliers = df_encoded[df_encoded['outlier'] == 1].drop(columns=['outlier'])
```

```python
# PPS (Predictive Power Score)
import ppscore as pps
```

```python
# Compute PPS scores
pps_scores = pps.matrix(df_no_outliers)
```
