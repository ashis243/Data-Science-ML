```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Cardiotocographic.csv')
```

```python
# Handle outliers (example: using IQR method)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum().sum()
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

```

```python
numerical_cols
```

```python
for col in numerical_cols :
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df
```

```python
df.isnull().sum().sum()
```

```python
# Data Visualization
# Histograms for numerical variables
df.hist(bins=20, figsize=(14, 10))
plt.tight_layout()
plt.show()
```

```python
# Boxplots to visualize distributions and potential outliers
plt.figure(figsize=(5, 3))
sns.boxplot(df=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation heatmap to see relationships between variables
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Pair plot for a deeper look into relationships between variables
sns.pairplot(df)
plt.show()
```

```python
# Replace NaNs with zeros (optional, depending on analysis needs)
correlation_matrix = correlation_matrix.fillna(0)

```

```python
correlation_matrix
```

```python
# Extract correlations above the threshold
high_correlation = correlation_matrix[abs(correlation_matrix) > 0.7]
print(high_correlation)
```
