```python
#change or convert categorical column into numeric column
df['Fuel_Type'] = df['Fuel_Type'].astype('category').cat.codes
```

```python
df
```

```python
#Split the Dataset
from sklearn.model_selection import train_test_split

# Define features (X) and target variable (y)
X = df.drop('Price', axis=1)
Y = df['Price']
```

```python
X
```

```python
Y
```

```python
# Split the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Build Multiple Linear Regression Models:
#Model 1: Basic Multiple Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model1 = LinearRegression()
model1.fit(X_train, Y_train)
```

```python
new_data = pd.DataFrame({
    'Age_08_04': [3, 5, 1],  # Example values for Age
    'KM': [15000, 30000, 5000],  # Example values for kilometers
    'Fuel_Type': [1, 0, 2],  # Replace with encoded values for Fuel_Type (e.g., CNG: 0, Diesel: 1, Petrol: 2)
    'HP': [90, 100, 70],  # Example values for horsepower
    'Automatic': [0, 1, 0],  # 0 for manual, 1 for automatic
    'cc': [1400, 1600, 1200],  # Example values for engine capacity
    'Doors': [4, 5, 4],  # Example values for number of doors
    'Gears': [5, 6, 5],  # Example values for number of gears
    'Weight': [1200, 1500, 1100]  # Example values for weight
})
```

```python
y_pred = model1.predict(new_data)
```

```python
print(y_pred)
```

```python
# Interpret coefficients
coefficients = pd.DataFrame(model1.coef_, X.columns, columns=['Coefficient'])
print(coefficients)
```

```python
#Model 2: Using Polynomial Features

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

X_train_poly, X_test_poly, Y_train_poly, Y_test_poly = train_test_split(X_poly, Y, test_size=0.2, random_state=42)

model2 = LinearRegression()
model2.fit(X_train_poly, Y_train_poly)
```

```python
new_data1 = pd.DataFrame({
    'Age_08_04': [4, 6],
    'KM': [17000, 40000],
    'Fuel_Type': [0, 2],  # Encoded values
    'HP': [80, 90],
    'Automatic': [1, 0],
    'cc': [1400, 1600],
    'Doors': [4, 5],
    'Gears': [5, 6],
    'Weight': [1200, 1500]
})

```

```python
new_data_poly = poly.transform(new_data1)
```

```python
y_pred = model2.predict(new_data_poly)
```

```python
print(y_pred)
```

```python
#Model 3: Using Feature Selection
from sklearn.feature_selection import SelectKBest, f_regression

# Adjust k to be <= the number of features in your dataset
selector = SelectKBest(score_func=f_regression, k=9)
X_selected = selector.fit_transform(X, Y)
X_train_sel, X_test_sel, Y_train_sel, Y_test_sel = train_test_split(X_selected, Y, test_size=0.2, random_state=42)
model3 = LinearRegression()
model3.fit(X_train_sel, Y_train_sel)
```

```python
new_data2 = pd.DataFrame({
    'Age_08_04': [5, 4],
    'KM': [18000, 30000],
    'Fuel_Type': [1, 0],  # Encoded values
    'HP': [70, 80],
    'Automatic': [0, 1],
    'cc': [1400, 1600],
    'Doors': [4, 5],
    'Gears': [5, 6],
    'Weight': [1200, 1500]
})
```

```python
new_data_selected = selector.transform(new_data2)
```

```python
y_pred = model3.predict(new_data_selected)
```

```python
print(y_pred)
```

```python
#Evaluate the Performance of the Models:
def evaluate_model(model, X_train, Y_train, X_test, Y_test):
    Y_train_pred = model.predict(X_train)
    Y_test_pred = model.predict(X_test)

    print("Training Performance")
    print("MSE:", mean_squared_error(Y_train, Y_train_pred))
    print("R2 Score:", r2_score(Y_train, Y_train_pred))

    print("\nTesting Performance")
    print("MSE:", mean_squared_error(Y_test, Y_test_pred))
    print("R2 Score:", r2_score(Y_test, Y_test_pred))

# Evaluate Model 1
evaluate_model(model1, X_train, Y_train, X_test, Y_test)

# Evaluate Model 2
evaluate_model(model2, X_train_poly, Y_train_poly, X_test_poly, Y_test_poly)

# Evaluate Model 3
evaluate_model(model3, X_train_sel, Y_train_sel, X_test_sel, Y_test_sel)
```

```python
#Apply Lasso and Ridge Regression:
from sklearn.linear_model import Lasso, Ridge

# Lasso Regression
lasso = Lasso(alpha=1.0)
lasso.fit(X_train, Y_train)
evaluate_model(lasso, X_train, Y_train, X_test, Y_test)

# Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, Y_train)
evaluate_model(ridge, X_train, Y_train, X_test, Y_test)
```
